```python
# let's find the cities around the Equator

equator_coordinates = list(filter(lambda coordinate: (-5 <= coordinate[0] <= 5), coordinates))


# using a function below is equivalent to the anonymous function or lambda above

def get_equator_cities(coordinate):
    near_equator = -5 <= coordinate[0] <= 5
    return near_equator

equatorial_coordinates = list(filter(get_equator_cities, coordinates))



#len(equator_coordinates)
len(equatorial_coordinates)
```

```python
#Get the nearest city using the citipy module.
from citipy import citipy
```

```python
# Create a list for holding the cities.
cities = []

set_cities = set()

# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    # city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # another way to do the same thing
    # if a method takes a tuple like coordinate as input
    # there is a special syntax called "unpacking the tuple" with *
    city = citipy.nearest_city(*coordinate).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
        
    # another way to do this is to use a set()
    # it will automatically remove duplicates. 
    set_cities.add(city)
    
# Print the city count to confirm sufficient count.
len(cities)

# convert set back to list
set_cities = list(set_cities)
len(set_cities)
```

```python
# Repeated for tropical cities

tropical_cities =  set()

# Identify the nearest city for each latitude and longitude combination.
for coordinate in equator_coordinates:
    # city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # another way to do the same thing
    # if a method takes a tuple like coordinate as input
    # there is a special syntax called "unpacking the tuple" with *
    city = citipy.nearest_city(*coordinate).city_name

    # another way to do this is to use a set()
    # it will automatically remove duplicates. 
    tropical_cities.add(city)
    
# Print the city count to confirm sufficient count.
len(tropical_cities)

# convert set back to list
tropical_cities = list(tropical_cities)
print(f"The {len(tropical_cities)} tropical cities are: \n\n{tropical_cities}")
```

```python
import time
# we'll try to wait to not exceed API limit of 60 / minute


# Create an empty list to hold the weather data.
city_data = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1


# Loop through all the cities in the list.
for i, city in enumerate(cities):

    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1
    
# Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_weatherdesc = city_weather["weather"][0]["description"]
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description": city_weatherdesc
                            })

# If an error is experienced, skip the city.
    except Exception as e:
        print("City not found. Skipping...", e)
        pass
    
    # sleep 1.2 seconds to make 50 API queries per minute < 60 limit 
    # 760 / 50 ~ 15 minutes
    # print("Sleeping for 1.2 seconds ")
    time.sleep(1.2)
    

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df.head()
```

```python
from urllib.parse import urlencode
import requests
import json
import time

# started using the wrong API at first
#https://developers.google.com/maps/documentation/directions/overview#directions-requests
    
direction_api_base_url = "https://maps.googleapis.com/maps/api/directions/json?"


# https://developers.google.com/places/web-service/search
place_api_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"


# 6a. Set parameters to search for hotels with 5000 meters.

params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

# this automatically adds the required "&" and "=" in the http URL query 
query = urlencode(params)

# 6b. Iterate through the hotel DataFrame.

for idx in hotel_df.index:

    # 6c. Get latitude and longitude from DataFrame
    
    # tup = hotel_df['Lat'][idx], hotel_df['Lng'][idx]
    # type(tup)
    # print(tup)
    
    # saved as "unpacked" tuple, in two separate variables
    # lat,lng = hotel_df['Lat'][idx], hotel_df['Lng'][idx]
    # print(lat,lng)


    # saved as "packed" or regular tuple: ('42.456, -22.123')
    location = hotel_df['Lat'][idx], hotel_df['Lng'][idx]
    
    
    #save the city name because we're using row index to iterate over data frame instead of labels/columns
    city = hotel_df['City'][idx]

    # 6d. Set up the base URL for the Google Directions API to get JSON data.

    # base_url was already defined above
    if not place_api_url:
        print("This won't run")
        place_api_url = "https://maps.googleapis.com/maps/api/directions/json?"
   

    # 6e. Make request and retrieve the JSON data from the search. 
    
    params['location'] = str(location).strip('()')
    #print(params)
    
    request_url = place_api_url + urlencode(params)
    
    # print(request_url)
    
    
    try:
        result = requests.get(request_url)
    except Exception as e:
        print(f"Error with the connection to places API: {e}\n")
        continue
    
    # check that request was OK == HTTP status code 200
    #
    # https://en.wikipedia.org/wiki/List_of_HTTP_status_codes

    if result.status_code != 200:
        print("HTTP request failed")
    
    
    # json.loads converts the row output above into a python dictionary
    result_dict = json.loads(result.content)
   
    # check that there was any hotel result or if not
    # skip to next for loop idx to check next location
    
    if not result_dict['results']:
        continue
    
    first_hotel = result_dict['results'][0]['name']
    #print(first_hotel)
    
    # 6f. Get the first hotel from the results and store the name, if a hotel isn't found skip the city.
    hotel_df.loc[hotel_df["City"] == city, 'Hotel Name'] =  first_hotel
    # test with only one API call, exit the for loop here
    time.sleep(0.2)
    
    print(hotel_df.loc[hotel_df["City"] == city])
```
